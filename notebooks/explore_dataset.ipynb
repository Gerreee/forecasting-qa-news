```python
import json
import pandas as pd
from pprint import pprint
```

# Load the dataset

```python
with open("../data/forecasting_qa_news.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
print(f"Total questions in dataset: {len(data)}")
```

```python
df = pd.json_normalize(data, sep='_')
```

# Explore the dataset   

```python
print("\nColumns in the dataset:")
print(df.columns.tolist())
```

```python
print("\nSample entry:")
pprint(data[0])
```

```python

print("\nUnique categories (flattened):")
from collections import Counter

all_categories = [cat for entry in data for cat in entry['categories']]
category_counts = Counter(all_categories)
for cat, count in category_counts.items():
    print(f"{cat}: {count}")

```

```python
# Minimum three news article per question
df['num_articles'] = df['news_articles'].apply(len)
print("\nNews articles per question (summary):")
print(df['num_articles'].describe())
```
